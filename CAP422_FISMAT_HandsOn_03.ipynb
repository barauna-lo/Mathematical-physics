Quantos	 padrões	 de	 Campo	 Vetorial,	 com	 a=b=1	 e	 x≠ 0	 e	 y≠ 0,	 são	 possíveis	 na	 aplicação:

$$ 
F:R^2 \rightarrow R^2 | V_F(x,y) = ax \hat{e}_1 + by\hat{e}_2
$$

Para $a=b=1$ e $x \ne 0$ e $y \ne 0$.

temos que a aplicação será

$$ 
F:R^2 \rightarrow R^2 | V_F(x,y) = x \hat{e}_1 + y\hat{e}_2
$$

Ou seja, esta aplicação irá gerar um campo vetorial radial com vetores em todas as direções. 

````python
import numpy as np
import matplotlib.pyplot as plt

x,y = np.meshgrid(np.linspace(-5,5,10) , np.linspace(-5,5,10))

u = x
v = y

plt.quiver(x,y,u,v)
plt.show()
````

Os tensores covariantes e contravariantes podem ser escritos, usando a notação utilizada em [astro.if.ufrgs](http://astro.if.ufrgs.br/evol/covar.htm), respectivamente como:

$$
T^{kl} =  \frac{\partial x_k}{\partial x_i} \frac{\partial x_l}{\partial x_j} T^{ij}
$$

$$
T_{kl} =  \frac{\partial x_i}{\partial x_k} \frac{\partial x_j}{\partial x_l} T_{ij}
$$


https://docs.sympy.org/latest/modules/tensor/tensor.html

````python
from sympy.tensor.tensor import TensorIndexType, TensorIndex, TensorHead, tensor_indices

Lorentz = TensorIndexType('Lorentz', dummy_name='L')

mu = TensorIndex('mu', Lorentz, is_up=False)

nu, rho = tensor_indices('nu, rho', Lorentz )

A = TensorHead('A', [Lorentz, Lorentz])
A(mu, nu)
A(-mu, -rho)
````

Seja um oscilador Hamonico simples, um sistema físico composto por um corpo de massa $m$ preso a um fio inextensível de massa depresível e comprimento $l$ sob ação de um campo gravitacional com aceleração $\vec{g}$, conforme mostra a figura abaixo.

A equação Lagrangiana é a diferença entre a energia cinética $T$ e a energia potêncial $U$ de um sistema físico. Escrevmos então 

$$
L = T-U.
$$

Para o problema do oscilador hamônico simplmes, temos que a energia cinética será 

$$
T = \frac{1}{2} m v^2 
$$

$$
T = \frac{1}{2} m \dot{x}^2 
$$

Se consideramos que a posição $x$ do oscilador hamonônico é descrito pelo arco de circunferência $l\theta$ temos que a energia cinética será

$$
T = \frac{1}{2} m r\dot{\theta}^2 
$$

Analogamente a energia potência gravitacional será 

$$
U = -mgh
$$
onde $h = l \cos(\theta)$ de forma que

$$
U = -mgl\cos(\theta)
$$

Assim a Lagrangiana desses sistema será 

$$
L = \frac{1}{2} m r\dot{\theta}^2  + mgl\cos(\theta)
$$

Vamos implmentar esta equação utilizando o ```sympy```

Existe o módulo [Classical Mechanics](https://docs.sympy.org/latest/modules/physics/mechanics/index.html) e o módulo [Lagrange’s Method](https://docs.sympy.org/latest/modules/physics/mechanics/lagrange.html) que precisam ser olhadas.

Dar uma olhada neste [link](https://stackoverflow.com/questions/25346132/is-it-possible-to-implement-newtons-dot-notation-or-lagranges-prime-notation) também

````python
from sympy import *
import numpy as np

L, T, U, m, r, theta, g, l, theta_t = symbols('L, T, U, m, r, theta, g, l, \dot{θ}')
````

````python
#Enegia Cinética
Eq(T,Integer(1)/2*m*l*theta_t**2)

#Energia Potêncial
Eq(U, -m*g*l*cos(theta))

#Lagrangiana
#Eq(L, T-U)
Eq(L, Integer(1)/2*m*l*theta_t**2 + m*g*l*cos(theta))
````
